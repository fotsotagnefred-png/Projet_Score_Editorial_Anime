```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Charger le fichier
df = pd.read_csv("animes.csv")

# 2. CLASSIFICATION : On crée une colonne Succès (1 si note >= 7, sinon 0)
df['Succes'] = (df['Note_Globale'] >= 7).astype(int)

# 3. LES DONNÉES : On choisit ce qu'on donne à l'IA pour apprendre
X = df[['Nb_Episodes', 'Note_Meilleur_Ep', 'Note_Pire_Ep']]
y = df['Succes']

# 4. LE TEST : On garde 20% des données pour tester l'ordinateur
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 5. LE MODÈLE : On crée et on entraîne l'IA
mon_modele = LogisticRegression()
mon_modele.fit(X_train, y_train)

# 6. MÉTRIQUE : On calcule le pourcentage de bonnes réponses (Accuracy)
devinettes = mon_modele.predict(X_test)
score = accuracy_score(y_test, devinettes)
print(f"La précision de mon IA est de : {score * 100:.2f} %")
# On demande à l'IA de deviner pour TOUS les 73 animes
df['IA_Devine'] = mon_modele.predict(X)
df.to_csv("resultat_fin.csv", index=False)

# Petit graphique simple pour l'export image
df['IA_Devine'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title("Nombre de Succès prédits par l'IA")
plt.savefig("mon_graphique_ia.png")
plt.show()
```
